```python
%matplotlib inline
# Code généré avec l'aide de l'IA de Google (2026)
from scipy.special import erf
import matplotlib.pyplot as plt
import numpy as np

# Définition des fonctions d'activation
def relu(x):
    return np.maximum(0, x)
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
def tanh(x):
    return np.tanh(x)
def swish(x):
    return x / (1 + np.exp(-x))
def gelu(x):
    return 0.5 * x * (1 + erf(x / np.sqrt(2)))
    
# Génération des données d'entrée
x = np.linspace(-4, 4, 1000)
# Tracé des courbes
plt.figure(figsize=(10, 6), dpi=100)
plt.plot(x, relu(x), label="ReLU", color="#ff7f0e", linewidth=2)
plt.plot(x, gelu(x), label="GELU", color="#2ca02c", linewidth=2.5)
plt.plot(x, swish(x), label="Swish / SiLU", color="#1f77b4", linestyle="--", linewidth=2)
plt.plot(x, sigmoid(x), label="Sigmoid", color="#d62728", linewidth=1.5)
plt.plot(x, tanh(x), label="Tanh", color="#9467bd", linewidth=1.5)

# Lignes de repère (axes x=0 et y=0)
plt.axhline(0, color="black", linewidth=0.8, linestyle="-", alpha=0.3)
plt.axvline(0, color="black", linewidth=0.8, linestyle="-", alpha=0.3)

# Personnalisation esthétique
plt.title("Comparaison des principales fonctions d'activation")
plt.xlabel("Entrée (x)", fontsize=11)
plt.ylabel("Sortie (f(x))", fontsize=11)
plt.xlim([-4, 4])
plt.ylim([-1.5, 3.5])
plt.grid(True, which="both", linestyle=":", alpha=0.5)
plt.legend(fontsize=11, loc="upper left")
plt.tight_layout()
plt.show()
```

```python
# Code généré avec l'aide de l'IA de Google (2026)
from scipy.special import erf
import matplotlib.pyplot as plt
import numpy as np

# --- Définition des Dérivées ---
def relu_derivative(x):
    return np.where(x > 0, 1.0, 0.0)
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
def sigmoid_derivative(x):
    s = sigmoid(x)
    return s * (1 - s)
def tanh_derivative(x):
    return 1 - np.tanh(x) ** 2
def swish_derivative(x):
    s = sigmoid(x)
    return s + x * s * (1 - s)
def gelu_derivative(x):
    # Formule exacte de la dérivée de GELU
    cdf = 0.5 * (1 + erf(x / np.sqrt(2)))
    pdf = np.exp(-(x**2) / 2) / np.sqrt(2 * np.pi)
    return cdf + x * pdf

# --- Génération du Graphique ---
x = np.linspace(-4, 4, 1000)
plt.figure(figsize=(11, 6), dpi=100)
# Tracé des courbes des gradients
plt.plot(x, relu_derivative(x), label="Gradient ReLU", color="#ff7f0e", linewidth=2)
plt.plot(x, gelu_derivative(x), label="Gradient GELU", color="#2ca02c", linewidth=2.5)
plt.plot(x,swish_derivative(x),label="Gradient Swish / SiLU",color="#1f77b4",linestyle="--",linewidth=2,)
plt.plot(x,sigmoid_derivative(x),label="Gradient Sigmoid",color="#d62728",linewidth=1.5,)
plt.plot(x, tanh_derivative(x), label="Gradient Tanh", color="#9467bd", linewidth=1.5)

# Mise en page et repères
plt.axhline(0, color="black", linewidth=0.8, linestyle="-", alpha=0.3)
plt.axvline(0, color="black", linewidth=0.8, linestyle="-", alpha=0.3)
plt.title("Comparaison des gradients (dérivées) des fonctions d'activation")
plt.xlabel("Entrée (x)", fontsize=11)
plt.ylabel("Gradient (f'(x))", fontsize=11)
plt.xlim([-4, 4])
plt.ylim([-0.2, 1.3])
plt.grid(True, which="both", linestyle=":", alpha=0.5)
plt.legend(fontsize=11, loc="upper left")
plt.tight_layout()
plt.show()
```
