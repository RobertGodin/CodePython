```python
# Simulation de AR(1) avec theta1 = 0.5
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline

np.random.seed(42)
n = 100
theta1 = 0.5
w = np.random.normal(size=n-1)
x = np.empty(n)
x[0] = np.random.normal()
for t in range(1,n):
    x[t] = theta1*x[t-1] + w[t-1]
plt.plot(x)
```

```python
from statsmodels.tsa.ar_model import AutoReg
modele_ar1 = AutoReg(x,lags=1)
modele_fit = modele_ar1.fit()
print('Modèle AR(1) : AIC = ', modele_fit.aic)
print(modele_fit.summary())
```

```python
modele_fit.plot_predict(start=2, end=len(x)+10)
plt.show()
```

```python
from statsmodels.tsa.ar_model import ar_select_order
modeles = ar_select_order(x,maxlag=10)
print("Décalages choisis:", modeles.ar_lags)
```
