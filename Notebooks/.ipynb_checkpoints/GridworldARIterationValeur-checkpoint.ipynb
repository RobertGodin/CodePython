```python
# -*- coding: utf-8 -*-
%matplotlib inline
import numpy as np
import random
import matplotlib.pyplot as plt
import matplotlib as mpl

def afficher_grille(grille):
    """ Afficher les cases de la grille.
    grille : tableau 2d de valeurs
    """
    largeur, hauteur = grille.shape
    plt.figure(figsize=(largeur, hauteur))
    im = plt.imshow(grille,interpolation='none',cmap = mpl.cm.binary)
    ax = plt.gca()
    # Graduation majeure pour les axes
    ax.set_xticks(np.arange(0, largeur, 1))
    ax.set_yticks(np.arange(0, hauteur, 1))
    # Etiquettes de graduation
    ax.set_xticklabels(np.arange(0, largeur, 1));
    ax.set_yticklabels(np.arange(0, hauteur, 1));
    # Graduations mineures
    ax.set_xticks(np.arange(-.5, largeur, 1), minor=True);
    ax.set_yticks(np.arange(-.5, hauteur, 1), minor=True);

    # Quadrillage pour délimiter les pixels
    ax.grid(which='minor', linestyle='-', linewidth=2)

    for i in range(largeur):
        for j in range(hauteur):
            text = ax.text(j, i,round(grille[i, j],2),ha="center", va="center",color="r")
            
    plt.show()

gamma = 1 # taux d'escompte
taille_grille = 4
recompense = -1
etats_terminaux = [[0,0], [taille_grille-1, taille_grille-1]]
actions = [[-1, 0], [1, 0], [0, 1], [0, -1]]
nombre_iterations = 1000

def p(s,a):
    """ Retourne l'état suivant s_prime et la récompense r"""
    if s in etats_terminaux:
        return s, 0
    s_prime = np.array(s) + np.array(a)
    r = recompense
    # si débordement de grille retour à l'état s
    if -1 in s_prime or taille_grille in s_prime:
        s_prime = s
    return s_prime, r

etats = [[i, j] for i in range(taille_grille) for j in range(taille_grille)]


def calculer_v_par_iteration(v,pi,etats,actions,nombre_iterations,gamma,seuil_delta = 0.001):
    """ Calcul de la récompense moyenne anticipée v pour la politique pi
    par itération de l'équation de Bellman """
    
    for i in range(nombre_iterations):
        copie_v=np.copy(v)
        delta = 0
        for etat in etats :
            v_nouveau=0
            for idx,action in enumerate(actions):
                s_prime,r = p(etat,action)
                v_nouveau += pi[etat[0],etat[1],idx]*(r + gamma * copie_v[s_prime[0],s_prime[1]])
            v[etat[0],etat[1]]=v_nouveau
            delta =max(delta,abs(v_nouveau-copie_v[etat[0],etat[1]]))
        
        if(delta < seuil_delta):
            break
    return i,delta,v
    
v = np.zeros((taille_grille, taille_grille)) 
pi = np.ones((taille_grille,taille_grille,4))/4

i,delta,v = calculer_v_par_iteration(v,pi,etats,actions,nombre_iterations,1)

print("delta:", delta, "iteration:",i)
print("-------------- v ---------------")
afficher_grille(v)
```

```python
def argmax_multi(a):
    if len(a) == 0:
        return []
    indices_max = [0]
    maximum = a[0]
    for i in range(1, len(a)):
        if a[i] > maximum:
            indices_max = [i]
            maximum = a[i]
        elif a[i] == maximum:
            indices_max.append(i)
    return indices_max

def vorace(etat,pi,gamma,v):
    """ Trouver meilleure politique pour un état """
    q_values=np.zeros(len(actions))
    for idx,action in enumerate(actions):
        s_prime,r = p(etat,action)
        q_values[idx] += 1/4* (r + gamma * v[s_prime[0],s_prime[1]])
    # Trouver l'indice de l'action dont la valeur de q est maximale 
    index_max =  np.argmax(q_values)
    for index_a in range(len(actions)):
        if index_a == index_max :
            pi[etat[0],etat[1],index_a] = 1
        else :
            pi[etat[0],etat[1],index_a] = 0
        
def ameliorer_pi_vorace(pi,gamma,v):
    politique_stable = True
    for etat in etats:
        old = pi[etat[0],etat[1]].copy()
        vorace(etat,pi,gamma,v)
        if not np.array_equal(pi[etat[0],etat[1]], old):
            politique_stable = False
    return pi,politique_stable

def iterer_evaluer_ameliorer(pi,etats,actions,nombre_iterations,gamma):
    """ Optimisation de la politique pi par itération de l'évaluation/amélioration vorace (policy iteration) """
    v = np.zeros((taille_grille, taille_grille))
    pi_pour_affichage = np.zeros((taille_grille, taille_grille))
    for i in range(nombre_iterations):
        nb_iterations,delta,v = calculer_v_par_iteration(v,pi,etats,actions,1,1)
        pi,politique_stable = ameliorer_pi_vorace(pi,gamma,v)
        if i in [0,1,2,10] or politique_stable:
            print("iteration du calcul de politique:",i)
            print("-------------- v ---------------")
            afficher_grille(v)
            print("------------ politique ------------ (0=haut,1=bas,2=droite,3=gauche)")
            for j in range(taille_grille):
                for k in range(taille_grille):
                    pi_pour_affichage[j,k]=np.argmax(pi[j,k])
            afficher_grille(pi_pour_affichage)
        if(politique_stable):
            print("politique stable itération :",i)
            break

# Initialiser v(s) à 0 et pi avec équiprobabilité des actions
v = np.zeros((taille_grille, taille_grille)) 
pi = np.ones((taille_grille,taille_grille,4))/4
            
iterer_evaluer_ameliorer(pi,etats,actions,nombre_iterations,1)
```
