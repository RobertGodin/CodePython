```python
%matplotlib inline
# -*- coding: utf-8 -*-
"""
Optimisation de politique avec mthode Q-learning
Politique epsilon-vorace
Environnement Taxi-v3 
"""
import gym
import numpy as np
import random
import itertools
import sys
import matplotlib
from matplotlib import pyplot as plt
random.seed(42)

def afficher_longueur_episode(longueur_episode):
    """
    Afficher l'volution des longueurs d'pisodes avec le temps
    """
    figure1 = plt.figure(figsize=(12,6))
    plt.plot(longueur_episode)
    plt.xlabel("pisode")
    plt.ylabel("Longueur pisode")
    plt.title("Longueur pisode par rapport au temps : Q-learning, Taxi-v3")
    
def afficher_retour_par_episode(retour_par_episode):
    """
    Afficher l'volution des rcompenses par pisode
    """
    figure1 = plt.figure(figsize=(12,6))
    plt.plot(retour_par_episode)
    plt.xlabel("pisode")
    plt.ylabel("Retour par pisode")
    plt.title("Retour par pisode par rapport au temps : Q-learning, Taxi-v3")

def politique_optimale_Q(env, nombre_episodes, max_t=99, gamma=1.0, alpha=0.1, epsilon=0.1):
    """
    Optimisation de politique avec mthode Q-learning
        env: environnement de type OpenAI gym
        nombre_episodes: nombre d'pisodes gnrs pour l'apprentissage
        gamma: facteur d'escompte des rcompenses futures
        alpha: taux d'apprentissage
        epsilon: probabilit pour politique e-vorace
    Retour:
        Q: np.array(nb_etats,nb_actions)
        retour_par_episode: tableau des retours par pisode
        longueur_episode: tableau des longueurs d'pisode
    """
    longueur_episode = np.zeros(nombre_episodes)
    retour_par_episode = np.zeros(nombre_episodes)
    nb_actions = env.action_space.n
    nb_etats = env.observation_space.n
    Q = np.zeros((nb_etats, nb_actions))
    
    for i_episode in range(nombre_episodes):
        if (i_episode+1) % 10 == 0:
            print("\rEpisode {}/{}.".format(i_episode, nombre_episodes), end="")
            sys.stdout.flush()

        etat = env.reset()
        for t in itertools.count():
            
            # Choisir action selon politique e-vorace
            unif_01 = random.uniform(0, 1)
            if unif_01 > epsilon:
                action = np.argmax(Q[etat,:])
            else:
                action = env.action_space.sample()
            
            etat_suivant, recompense, final, _ = env.step(action)
            Q[etat,action] = Q[etat,action]+alpha*(recompense+gamma*np.max(Q[etat_suivant,:])-Q[etat,action])
            retour_par_episode[i_episode]+=recompense
            if final:
                longueur_episode[i_episode] = t
                break
            etat = etat_suivant
        
    return Q,retour_par_episode,longueur_episode

env = gym.make("Taxi-v3")
env.render()

Q,retour_par_episode,longueur_episode = politique_optimale_Q(env, nombre_episodes=5000,alpha=0.1)
afficher_retour_par_episode(retour_par_episode)
afficher_longueur_episode(longueur_episode)
```

```python
""" Test du jeu aprs optimisation de Q """
nb_episodes = 3
max_etapes = 99 
env.reset()
retour_par_episode = []

for episode in range(nb_episodes):
    etat = env.reset()
    final = False
    total_recompenses = 0
    print("**********************************")
    print("pisode : ", episode)

    for t in range(max_etapes):
        env.render()
        action = np.argmax(Q[etat])
        
        etat_suivant, recompense, final, _ = env.step(action)
        
        total_recompenses += recompense
        
        if final:
            retour_par_episode.append(total_recompenses)
            print ("Total des rcompenses de l'pisode :", total_recompenses)
            print("Longueur pisode :", t)
            env.render()
            break
        etat = etat_suivant
env.close()
print ("Moyenne de rcompenses par pisode: " +  str(sum(retour_par_episode)/nb_episodes))
```
