```python
# Chargement des données de Iris 
# x : nombre de pieces y : prix 

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import LogisticRegression

iris = datasets.load_iris()
print("----------------- Description ------------------")
print(iris.DESCR)
print("----------------- iris.keys ------------------")
print(list(iris.keys()))
# print("----------------- iris ------------------")
# print(iris)
print("----------------- Variables indépendantes (X) 150 lignes par 4 colonnes : lignes 0:5 -----------------")
print(iris.data[0:5])
print("----------------- Variable à prédire (y) : type d'Iris, lignes 0:5 -----------------")
print(iris.target[0:5])

print("Colonne largeur de pétale dans un tableau 2D, lignes 0:5")
iris_X = iris.data[:, np.newaxis, 3] # X est la colonne largeur de pétale
print(iris_X[0:5])
print("X")

iris_y = iris.target

plt.scatter(iris_X,iris_y)
plt.title("Classe d'Iris vs largeur de pétale")
plt.xlabel('largeur de pétale')
plt.ylabel('classe')
plt.legend(loc='lower right')
plt.show()

X = iris["data"][:,3:]  # petal width
y = (iris["target"]==2).astype(np.int)

log_reg = LogisticRegression(penalty="l2")
log_reg.fit(X,y)

X_new = np.linspace(0,3,1000).reshape(-1,1)
y_proba = log_reg.predict_proba(X_new)

plt.plot(X,y,"b.")
plt.plot(X_new,y_proba[:,1],"g-",label="Iris-Virginica")
plt.plot(X_new,y_proba[:,0],"b--",label="Not Iris-Virginca")
plt.xlabel("Petal width", fontsize=14)
plt.ylabel("Probability", fontsize=14)
plt.legend(loc="upper left", fontsize=14)
plt.show()
```
