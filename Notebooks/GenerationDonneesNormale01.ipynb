```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
def densite_normale(x,mu=0,sigma=1):
    return 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (x - mu)**2 / (2 * sigma**2))
mu, sigma = 0, 1
np.random.seed(22) # Pour obtenir des résultats reproductibles avec les données aléatoires
Y = np.random.normal(mu, sigma, 100000)
comptes, frontieres, _ = plt.hist(Y, 50, density=True)
plt.plot(frontieres, densite_normale(frontieres,mu=0,sigma=1))
plt.show()
```
