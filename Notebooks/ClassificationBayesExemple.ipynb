```python
# Importation des bibliothèques nécessaires
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import CategoricalNB
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
import numpy as np

# Variables sexe,âge et revenu
X = np.array(
[
    ["m",	"jeune",	"faible"],
    ["m",	"jeune",	"élevé"],
    ["f",	"jeune",	"élevé"],
    ["m",	"jeune",	"faible"],
    ["f",	"vieux",	"élevé"],
    ["f",	"vieux",	"faible"],
    ["f",	"jeune",	"faible"],
    ["m",	"vieux",	"élevé"],
    ["m",	"vieux",	"faible"],
    ["f",	"vieux",	"faible"],
])

# Oui pour joueur
y_text = np.array(["non", "oui", "oui", "oui", "non", 
                   "non", "non", "oui", "non", "non"])

# Encodage des variables prédictives sous forme d'entiers pour le classifieur CategoricalNB()
encodeur_X = LabelEncoder()
X_encode = np.array([encodeur_X.fit_transform(v) for v in X.T]).T
print("X encodé:")
print(X_encode)

# Encodage des étiquettes textuelles
encodeur_y = LabelEncoder()
y_encode = encodeur_y.fit_transform(y_text)
print("y encodé:")
print(y_encode)

# Création du classifieur bayésien naïf pour les données catégoriques
classifieur = CategoricalNB()

# Entraînement du modèle sur les données complètes
classifieur.fit(X_encode, y_encode)

# Prédictions pour toutes les données
y_pred_encode = classifieur.predict(X_encode)
print("y_pred_encode:",y_pred_encode)

# Évaluation de la précision du modèle
accuracy = accuracy_score(y_encode, y_pred_encode)
print(f"Précision du modèle : {accuracy:.2f}")

# Décodage des étiquettes prédites
y_pred_text = encodeur_y.inverse_transform(y_pred_encode)

# Affichage des résultats de prédiction
print("Étiquettes réelles : ", encodeur_y.inverse_transform(y_encode))
print("Étiquettes prédites : ", y_pred_text)
```

```python
# Importation des bibliothèques nécessaires
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import CategoricalNB
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
import numpy as np

# Variables sexe,âge et revenu
X = np.array(
[
    ["m",	"jeune",	"faible"],
    ["m",	"jeune",	"élevé"],
    ["f",	"jeune",	"élevé"],
    ["m",	"jeune",	"faible"],
    ["f",	"vieux",	"élevé"],
    ["f",	"vieux",	"faible"],
    ["f",	"jeune",	"faible"],
    ["m",	"vieux",	"élevé"],
    ["m",	"vieux",	"faible"],
    ["f",	"vieux",	"faible"],
])

# Oui pour joueur
y_text = np.array(["non", "oui", "oui", "oui", "non", 
                   "non", "non", "oui", "non", "non"])

# Encodage des variables prédictives sous forme d'entiers pour le classifieur CategoricalNB()
encodeur_X = LabelEncoder()
X_encode = np.array([encodeur_X.fit_transform(v) for v in X.T]).T
print("X encodé:")
print(X_encode)

# Encodage des étiquettes textuelles
encodeur_y = LabelEncoder()
y_encode = encodeur_y.fit_transform(y_text)
print("y encodé:")
print(y_encode)

# Séparation des données en ensembles d'entraînement et de test
X_ent, X_test, y_ent, y_test = train_test_split(X_encode, y_encode, test_size=0.3, random_state=42)

print("X entrainement encodé:",X_ent)
print("y entrainement encodé:", y_ent)
print("X test encodé:",X_test)
print("y test encodé:", y_test)

# Création du classifieur bayésien naïf pour les données catégoriques
classifieur = CategoricalNB()

# Entraînement du modèle sur les données d'entraînement
classifieur.fit(X_ent, y_ent)

# Prédiction des étiquettes pour les données de test
y_pred = classifieur.predict(X_test)
print("y_pred:",y_pred)

# Évaluation de la précision du modèle
accuracy = accuracy_score(y_test, y_pred)
print(f"Précision du modèle : {accuracy:.2f}")

# Décodage des étiquettes prédites
y_pred_text = encodeur_y.inverse_transform(y_pred)

# Affichage des résultats de prédiction
print("Étiquettes réelles : ", encodeur_y.inverse_transform(y_test))
print("Étiquettes prédites : ", y_pred_text)
```
