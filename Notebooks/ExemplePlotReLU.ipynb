```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def f1(x):
    return 20+4*x
def f2(x):
    return -3*x
def f3(x):
    return -10+2*x

def afficher_graphique(f,x_inf, x_sup, titre):
    serie_x = np.linspace(x_inf,x_sup,100)
    serie_y = f(serie_x)
    plt.plot(serie_x,serie_y)
    plt.title(titre)
    plt.show()

afficher_graphique(f1,-10,10,"f1")
afficher_graphique(f2,-10,10,"f2")
afficher_graphique(f3,-10,10,"f3")
```

```python
def relu(x):
    return np.maximum(x,0)
def relu_f1(x):
    return relu(f1(x))
def relu_f2(x):
    return relu(f2(x))
def relu_f3(x):
    return relu(f3(x))

afficher_graphique(relu_f1,-10,10,"relu(f1)")
afficher_graphique(relu_f2,-10,10,"relu(f2)")
afficher_graphique(relu_f3,-10,10,"relu(f3)")
```

```python
def resultat(x):
    return relu(f1(x))+relu(f2(x))+relu(f3(x))
afficher_graphique(resultat,-10,10,"Résultat")
```
