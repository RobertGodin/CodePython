```python
# Algorithme de descente du gradiant pour x**2 avec taux = 0.1 et graphique de la tangente 

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

x = 3
nb_iterations = 15
taux = 0.1

liste_x=[]
liste_y=[]
for i in range(nb_iterations) :
    y = x**2
    liste_x.append(x)
    liste_y.append(y)
    x = x-taux*2*x

x=np.arange(-3.9,4,0.1)
plt.plot(x,x**2,label = "y=x*x")
plt.scatter(liste_x,liste_y,label="descente de gradiant avec taux="+str(taux))
plt.plot([2,4],[3,15],label = "tangente à x=3")
plt.title("Descente de gradiant pour y = x*x")
plt.legend(loc='upper center')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
plt.scatter(range(nb_iterations),liste_y,label="valeurs de y")
plt.title("Evolution de la valeur de y")
plt.xlabel('itération')
plt.ylabel('y')
plt.show()
```

```python
# Algorithme de descente du gradiant avec seuil de précidion
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

x = 3
y = 3**2
nb_iterations = 1000
taux = 0.1
seuil = 0.001
liste_x=[x]
liste_y=[y]

for i in range(nb_iterations) :
    x_suivant = x-taux*2*x
    y_suivant = x_suivant**2
    liste_x.append(x_suivant)
    liste_y.append(y_suivant)
    if abs(y_suivant-y) < seuil :
        print("Nombre d'itérations :", i)
        print("x:",x_suivant," y:",y_suivant)
        break
    else :
        x = x_suivant
        y = y_suivant

x=np.arange(-3.9,4,0.1)
plt.plot(x,x**2,label = "y=x*x")
plt.scatter(liste_x,liste_y,label="descente gradiant avec taux="+str(taux))
plt.title("Descente de gradiant pour y = x*x")
plt.legend(loc='upper center')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
plt.scatter(range(len(liste_y)),liste_y,label="valeurs de y")
plt.title("Evolution de la valeur de y")
plt.xlabel('itération')
plt.ylabel('y')
plt.show()
```
