```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
def densite_normale(x,mu=0,sigma=1):
    return 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (x - mu)**2 / (2 * sigma**2))
X = np.arange(-10,10,0.01) 
plt.plot(X, densite_normale(X,mu=0,sigma=1),label = 'Sigma=1')
plt.plot(X, densite_normale(X,mu=0,sigma=2),label = 'Sigma=2')
plt.plot(X, densite_normale(X,mu=0,sigma=3),label = 'Sigma=3')
plt.legend(loc='upper right')
plt.show()
```
