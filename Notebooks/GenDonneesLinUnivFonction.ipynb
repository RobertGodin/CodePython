```python
# Génération de données qui suivent une distribution linéaire univariée avec résidus ~ N(0,1)
# Version avec une fonction generer_donnees_linaire_univarie()
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

def h_lineaire_univarie(theta,x):
    """ Retourner la valeur de la fonction linéaire"""
    return theta[0]+theta[1]*x

def generer_donnees_linaire_univarie(m,theta,germe=22):
    """ Génération de données qui suivent une distribution linéaire univariée avec résidus ~ N(0,1) 
    
    m: nombre de lignes
    theta: n-uplet de la forme (theta0,theta1)
    germe: germe pour la génération des nombres aléatoires (defaut 22)
    
    retourne
    
    X: vecteur numpy de taille m de valeurs entre 0 et m
    Y: vecteur numpy de taille m avec les valeurs de y correspondantes
    """
    
    m=10 # nombre de lignes de données
    theta_simule = (5,1)
    np.random.seed(22) # Pour obtenir des résultats reproductibles avec les données aléatoires

    X = np.random.rand(m)*m # m valeurs de x entre 0 et m
    Y = h_lineaire_univarie(theta_simule,X)+np.random.randn(m) # Résidus N(0,1)
    return X,Y

m=10
theta_simule=(5,1) 
X,Y = generer_donnees_linaire_univarie(m,theta_simule)
plt.scatter(X,Y,label = 'Points simulés')

X_modele = np.arange(0,m+1,m)
Y_modele = h_lineaire_univarie(theta_simule,X_modele)
plt.plot(X_modele,Y_modele, '-r',label = 'Droite à simuler')

plt.title("Données générées pour régression linéaire y=5+x")
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='lower right')
plt.show()
```
