```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

x = 3
nb_iterations = 15
vitesse = 0.1

liste_x=[]
liste_y=[]
for i in range(nb_iterations) :
    y = x**2
    liste_x.append(x)
    liste_y.append(y)
    x = x-vitesse*2*x

x=np.arange(-3.9,4,0.1)
plt.plot(x,x**2,label = "y=x*x")
plt.scatter(liste_x,liste_y,label = "descente gradiant")
plt.plot([2,4],[3,15],label = "tangente à x=3")
plt.legend(loc='upper center')
plt.show()
```
