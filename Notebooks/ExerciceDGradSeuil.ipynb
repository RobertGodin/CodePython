```python
# Algorithme de descente du gradient avec seuil de pr√©cision
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

x = 3
y = 3**2
nb_iterations = 1000
taux = 0.1
seuil = 0.001
liste_x=[x]
liste_y=[y]

for i in range(nb_iterations) :
    x_suivant = x-taux*2*x
    y_suivant = x_suivant**2
    liste_x.append(x_suivant)
    liste_y.append(y_suivant)
    if abs(y_suivant-y) < seuil :
        print("Nombre d'it√©rations :", i)
        print("x:",x_suivant," y:",y_suivant)
        break
    else :
        x = x_suivant
        y = y_suivant

x=np.arange(-3.9,4,0.1)
plt.plot(x,x**2,label = "y=x*x")
plt.scatter(liste_x,liste_y,label="descente gradient avec taux="+str(taux))
plt.title("Descente de gradient pour y = x*x")
plt.legend(loc='upper center')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
plt.scatter(range(len(liste_y)),liste_y,label="valeurs de y")
plt.title("Evolution de la valeur de y")
plt.xlabel('it√©ration')
plt.ylabel('y')
plt.show()
```
