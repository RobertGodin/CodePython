```python
# Chargement des données de California Housing pour exercice de régression
# x : revenu y : prix 

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from math import sqrt
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score

from sklearn.datasets import fetch_california_housing
maisons_californie = fetch_california_housing()

print("----------------- Variables indépendantes (X) 20 640 lignes par 8 colonnes : lignes 0:5 -----------------")
print(maisons_californie.data[0:5])
print("----------------- Variable à prédire (y) : prix en $100 000, lignes 0:5 -----------------")
print(maisons_californie.target[0:5])

print("Colonne MedInc (revenu médian) dans un tableau 2D, lignes 0:5")
print(type(maisons_californie.data))
maisons_X = maisons_californie.data[:, np.newaxis, 0][0:500] # X est la colonne du revenu médian
print(maisons_X[0:5])
maisons_y = maisons_californie.target[0:500]

plt.scatter(maisons_X,maisons_y)
plt.title('Données du prix vs le revenu pour California Housing')
plt.xlabel('revenu')
plt.ylabel('prix')
plt.show()

# Estimation des paramètres avec le modèle de régression linéaire
modele = linear_model.LinearRegression()
modele.fit(maisons_X, maisons_y)

# Prédictions avec les observations
maisons_y_estime = modele.predict(maisons_X)

print("Theta0:", modele.intercept_)
print("Theta1:", modele.coef_[0])
print("Erreur quadratique moyenne :", mean_squared_error(maisons_y_estime,maisons_y))

plt.scatter(maisons_X,maisons_y)
plt.plot(maisons_X,maisons_y_estime,label='Droite estimée')

plt.title('Régression linéaire California Housing')
plt.xlabel('revenu')
plt.ylabel('prix')
plt.legend(loc='lower right')
plt.show()
```
